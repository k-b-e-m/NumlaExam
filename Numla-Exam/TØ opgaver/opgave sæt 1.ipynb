```python
print("opgave 1.5")
import matplotlib.pyplot as plt
import numpy as np
x = np.array([ 0.2, 0.2, -0.2, 0.3])
y = np.array([-0.2, 0.8, 0.1, 0.1])
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(x,y)
print("A")
a = 2.0
b = -1.0
c = -1.0
d = 2.0
ax.plot(a*x+b*y, c*x+d*y)

```

```python
print("1.5 B")
fig, bx = plt.subplots()
bx.set_aspect('equal')
bx.plot(x,y)
a = 0.5
b = -0.86
c = 0.86
d = 0.5
print("Første rotation")
bx.plot(a*x+b*y, c*x+d*y)
print("anden rotation")
a = 0.1
b = -np.sqrt((1-a))
c = np.sqrt((1-a))
d = 0.1
fig, cx = plt.subplots()
cx.set_aspect('equal')
cx.plot(x,y)
cx.plot(a*x+b*y, c*x+d*y)
print("Derved er en rotation efterfulgt af en anden rotation en rotation")
```

```python
print("1.5 c")
fig, bx = plt.subplots()
bx.set_aspect('equal')
bx.plot(x,y)
C =0.3
S = np.sqrt(1-(C*C))
a = C
b = S
c = S
d = -C
print("Første spejlning")
bx.plot(a*x+b*y, c*x+d*y)
print("anden spejlning")
C =0.9
S = np.sqrt(1-(C*C))
print("C²+S² =1:",(C*C) +S*S == 1)
a = C
b = S
c = S
d = -C
fig, cx = plt.subplots()
cx.set_aspect('equal')
cx.plot(x,y)
cx.plot(a*x+b*y, c*x+d*y)
print("Derved er en spejlning efterfulgt af en anden spejlning en Spejlning")
```
