# Opgave 4.1.
Lad $A \in \mathbb{R}^{1000 \times 70}, B, C \in \mathbb{R}^{70 \times 3000}, D \in \mathbb{R}^{3000 \times 1000}$. Angiv antallet af flops for de følgende matrixberegninger: 

(a) $A B$, 

(b) $A(B+C)$, 

(c) (CD) $A$, 

(d) $C(D A)$, 

(e) $(A(B D)) A$, 

(f) $(A B)(D A)$, 

(g) $A(B(D A))$.

```python
import numpy as np
A = np.array([1000,70])
B = np.array([70,3000])
C = np.array([70,3000])
D = np.array([3000,1000])
CA = np.array([C[0],A[1]])
DA = np.array([D[0],A[1]])
BD = np.array([B[0],D[1]])
ABD = np.array([A[0],BD[1]])
AB = np.array([A[0],B[1]])
BDA = np.array([B[0],DA[1]])

def computeSizeOfMatrix(side):
    return eval(side)
    

def computeFlops(expression):
        exp = expression.split()
        op = exp[1]
        left = exp[0]
        right = "".join(exp[2:])
        match op:
            case "+":
                return computeSizeOfMatrix(left)[0] * computeSizeOfMatrix(left)[1] #matrix need to be same size
            
            case "-":
                return computeSizeOfMatrix(left)[0] * computeSizeOfMatrix(left)[1] #matrix need to be same size
                
            case "@":
                return 2*computeSizeOfMatrix(left)[0] * computeSizeOfMatrix(right)[0]* computeSizeOfMatrix(right)[1]
            

print("a",computeFlops("A @ B"))
print("b",computeFlops("A @ B") + computeFlops("B + C"))
print("c", computeFlops("C @ A") + computeFlops("A @ CA"))
print("d",computeFlops("D @ A") + computeFlops("C @ DA"))
print("e",computeFlops("B @ D") + computeFlops("BD @ A") + computeFlops("ABD @ A"))
print("f",computeFlops("A @ B") + computeFlops("D @ A") + computeFlops("AB @ DA"))
print("g",computeFlops("D @ A") + computeFlops("DA @ B") + computeFlops("A @ BDA"))
```
